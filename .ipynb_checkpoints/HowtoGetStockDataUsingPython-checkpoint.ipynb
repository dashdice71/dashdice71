```python
import yfinance as yf

#define the ticker symbol
tickerSymbol = 'MSFT'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
tickerDf = tickerData.history(period='1d', start='2010-1-1', end='2021-12-25')

#see your data
tickerDf
```

```python
"""
info

For many tickers, you can get valuable information on the company as 
a whole such as geographic location, website, business summary, etc.

"""
#define the ticker symbol
tickerSymbol = 'MSFT'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#info on the company
tickerData.info
```

```python
"""
calendar

You can also use the yfinance library to get upcoming earnings events regarding the company.
"""

#define the ticker symbol
tickerSymbol = 'MSFT'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get event data for ticker
tickerData.calendar
```

```python
"""
recommendations

Finally, you can use the recommendations attribute to see how analysts have been rating this stock.
"""
#define the ticker symbol
tickerSymbol = 'MSFT'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get recommendation data for ticker
tickerData.recommendations
```
